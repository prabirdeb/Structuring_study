```python
def concept_data(k, student_data):
  # Getting google sheet id
  gsheetid = student_data[(student_data["ID"]==k)]["Concept_link"].values[0].replace("https://docs.google.com/spreadsheets/d/","").split("/")[0]
  sheet_name = "Concepts" # Student should not change the sheet name

  # Converting google sheet to csv
  gsheet_url = "https://docs.google.com/spreadsheets/d/{}/gviz/tq?tqx=out:csv&sheet={}".format(gsheetid, sheet_name)

  # Creating individual student data df
  individual_student_data = pd.read_csv(gsheet_url)
  individual_student_data = individual_student_data[(individual_student_data["Class"]==student_data[(student_data["ID"]==k)]["Class"].values[0])]
  return individual_student_data
```

```python
# calculating concepts learnt per subject per day (expected minimum 5)
def metric1(report, subject_data, day, sub):
  if report.Total_concept.shape[0]==0:
    if subject_data.shape[0]<=5:
      metric_1=subject_data.shape[0]*100/5
    else:
      metric_1=100
  else:
    if (subject_data.shape[0]-report[(report["Day"]==day-1) & (report["Subjects"]==sub)]["Total_concept"].values[0])<=5:
      metric_1=(subject_data.shape[0]-report[(report["Day"]==day-1) & (report["Subjects"]==sub)]["Total_concept"].values[0])*100/5
    else:
      metric_1=100
  return metric_1
```

```python
# Calculating increase in avg_words per concept per day (expected 2 per day per concept to ensure revisiting of the concepts)
def metric2(report,avg_words,day,sub):
  if report.Total_image.shape[0]==0:
    if avg_words<=50:
      metric_2=avg_words*100/50
    else:
      metric_2=100
  else:
    if (avg_words-report[(report["Day"]==day-1) & (report["Subjects"]==sub)]["Avg_word_per_concept"].values[0])<=2:
      metric_2=(avg_words-report[(report["Day"]==day-1) & (report["Subjects"]==sub)]["Avg_word_per_concept"].values[0])*100/2
    else:
      metric_2=100
  return metric_2
  
```

```python
# Calculating image explored per subject per day (expected minimum 3)
def metric3(report,image,day,sub):
  if report.Total_image.shape[0]==0:
    if image<=3:
      metric_3=image*100/3
    else:
      metric_3=100
  else:
    if (image-report[(report["Day"]==day-1) & (report["Subjects"]==sub)]["Total_image"].values[0])<=3:
      metric_3=(image-report[(report["Day"]==day-1) & (report["Subjects"]==sub)]["Total_image"].values[0])*100/3
    else:
      metric_3=100
  
  return metric_3
```

```python
# Calculating chapter read per subject per day (expected 1 chapter per day)
def metric4(report, count_chapter,day,sub):
  if report.Chapters_read.shape[0]==0:
    if count_chapter<=1:
      metric_4=count_chapter*100
    else:
      metric_4=100
  else:
    if (count_chapter-report[(report["Day"]==day-1) & (report["Subjects"]==sub)]["Chapters_read"].values[0])<=1:
      metric_4=(count_chapter-report[(report["Day"]==day-1) & (report["Subjects"]==sub)]["Chapters_read"].values[0])*100
    else:
      metric_4=100
  return metric_4
```

```python
# Calculating the average score to understand the quality of learning (expected above 75%)

def metric5(eval_PD,eval_SC,eval_SK,k):
  # getting the latest score given by prabir
  eval_PD_clean=eval_PD[(eval_PD["ID"]==k)][["score-a","score-b","score-c","score-d","score-e","score-f","score-g","score-h","score-i","score-j","score-k","score-l","score-m","score-n","score-o","score-p","score-q","score-r","score-s","score-t","score-u","score-v","score-w","score-x"]]
  eval_PD_clean.dropna(axis=1, inplace=True)
  if eval_PD_clean.shape[1]==0:
    pd=7.5
  else:
    pd=eval_PD_clean.iloc[:,-1][0]
  
  # getting the latest score given by shashwata
  eval_SC_clean=eval_SC[(eval_SC["ID"]==k)][["score-a","score-b","score-c","score-d","score-e","score-f","score-g","score-h","score-i","score-j","score-k","score-l","score-m","score-n","score-o","score-p","score-q","score-r","score-s","score-t","score-u","score-v","score-w","score-x"]]
  eval_SC_clean.dropna(axis=1, inplace=True)
  if eval_SC_clean.shape[1]==0:
    sc=7.5
  else:
    sc=eval_SC_clean.iloc[:,-1][0]
  
  # getting the latest score given by Sudip
  eval_SK_clean=eval_SK[(eval_SK["ID"]==k)][["score-a","score-b","score-c","score-d","score-e","score-f","score-g","score-h","score-i","score-j","score-k","score-l","score-m","score-n","score-o","score-p","score-q","score-r","score-s","score-t","score-u","score-v","score-w","score-x"]]
  eval_SK_clean.dropna(axis=1, inplace=True)
  if eval_SK_clean.shape[1]==0:
    sk=7.5
  else:
    sk=eval_SK_clean.iloc[:,-1][0]
  
  # Calculating metric-5
  metric_5=(pd+sc+sk)/3
  
  return metric_5
```
